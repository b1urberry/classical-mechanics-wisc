## 1 Lagrange Equatiions with a Nonconservative Force

Let $x$ represent the horizontal motion and $y$ represent the vertical motion.

Kinetic energy:

$$
T = \frac{1}{2}m(\dot{x}^2+\dot{y}^2)
$$

Potential energy:

$$
U = mgy
$$

Lagrangian:

$$
\mathcal{L} = T- U = \frac{1}{2}m(\dot{x}^2+\dot{y}^2) - mgy
$$

Quadratic drag force:

$$
\begin{aligned}
    \vec{f} &= -c\dot{v}^2\hat{v}\\
    f_x & = -c\sqrt{\dot{x}^2+\dot{y}^2}\dot{x}\\
    f_y & = -c\sqrt{\dot{x}^2+\dot{y}^2}\dot{y}\\ 
\end{aligned} 
$$

where $c$ is the quadratic drag coefficient.

There are two degrees of freedom, thus two generalized coordinates, $x$ and $y$.

Generalized drag force:

$$
\begin{aligned}
    Q_x &= f_x\frac{\partial x}{\partial x} + f_y\frac{\partial y}{\partial x} = f_x\\\\
    Q_y &= f_x\frac{\partial x}{\partial y} + f_y\frac{\partial y}{\partial y} = f_y
\end{aligned}
$$

Since quadratic drag force is nonconservative, we derive the equation of motion through the following:

Equation of motion for $x$ coordinate:

$$
\begin{aligned}
    \frac{d}{dt}\frac{\partial \mathcal{L}}{\partial \dot{x}} - \frac{\partial \mathcal{L}}{\partial x} &= Q_x\\\\
    m\ddot{x} &= -c\sqrt{\dot{x}^2+\dot{y}^2}\dot{x}\\
\end{aligned}
$$

Equation of motion for $y$ coordinate:

$$
\begin{aligned}
    \frac{d}{dt}\frac{\partial \mathcal{L}}{\partial \dot{y}} - \frac{\partial \mathcal{L}}{\partial y} &= Q_y\\\\
    m\ddot{y} &= -c\sqrt{\dot{x}^2+\dot{y}^2}\dot{y} - mg\\
\end{aligned}
$$

The results are consistent with what we derived using Newton's Law.


## 2 Conservation of Angular Momentum

### Question a

For the virtual displacement

$$
(r_\alpha, \theta_\alpha, \phi_\alpha) \rightarrow (r_\alpha, \theta_\alpha, \phi_\alpha + \epsilon)
$$

the potential energy is unchanged under rotation:

$$
U(r_\alpha, \theta_\alpha, \phi_\alpha) = U(r_\alpha, \theta_\alpha, \phi_\alpha + \epsilon)
$$

so, $\delta U = 0$ with the virtual displacement $\epsilon$.

The kinetic energy:

$$
T = \sum_{\alpha}\frac{1}{2}m_\alpha(\dot{r}_\alpha^2 + r_\alpha^2\dot{\theta}_\alpha^2 + r_\alpha^2\sin^2\theta_\alpha\dot{\phi}_\alpha^2)
$$

With $\frac{d}{dt}(\phi_\alpha + \epsilon) = \dot{\phi}_\alpha$ and $\delta r_\alpha = 0$ and $\delta \theta_\alpha = 0$, we know that the kinetic energy is unchanged with the virtual displacement $\epsilon$, i.e. $\delta T = 0$.

With $\delta T = 0$ and $\delta U = 0$, we have $\delta \mathcal{L} = 0$

For an $N$-particle system, consider a virtual displacement $\epsilon$ on the $\phi$ coordinate:

$$
\delta \mathcal{L} = \epsilon\frac{\partial \mathcal{L}}{\partial \phi_1} + \epsilon\frac{\partial \mathcal{L}}{\partial \phi_2} + ... + \epsilon\frac{\partial \mathcal{L}}{\partial \phi_N} = \epsilon\sum_{\alpha=1}^N \frac{\partial \mathcal{L}}{\partial \phi_\alpha}= 0
$$

With $\epsilon \neq 0$, we have:

$$
\sum_{\alpha=1}^N \frac{\partial \mathcal{L}}{\partial \phi_\alpha}= 0\tag{1}
$$

### Question b

According to the Euler-Lagrange equation for $\phi$ coordinate:

$$
\frac{\partial\mathcal{L}}{\partial\phi_\alpha} = \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\phi}_\alpha}
$$

Thus, with $(1)$ we have

$$
\begin{aligned}
    0 &= \sum_{\alpha=1}^N \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\phi}_\alpha} = \sum_{\alpha=1}^N \frac{\partial \mathcal{L}}{\partial \phi_\alpha}\\\\
     &= \frac{d}{dt}\sum_{\alpha=1}^N\frac{\partial\mathcal{L}}{\partial\dot{\phi}_\alpha} \\\\
     &= \frac{d}{dt}\sum_{\alpha=1}^N r_\alpha^2\sin^2\theta_\alpha\dot{\phi}_\alpha\\\\
0\cdot    m_\alpha &= \frac{d}{dt}\sum_{\alpha=1}^N (r_\alpha\sin\theta_\alpha)(r_\alpha\sin\theta_\alpha\dot{\phi}_\alpha m_\alpha)\\\\
     &= \frac{d}{dt}\sum_{\alpha=1}^N |\vec{r}_\alpha\times\vec{p}_\alpha|\\\\
     &= \frac{d}{dt}\sum_{\alpha=1}^N l_\alpha\\\\
     & = \frac{dL_z}{dt}
\end{aligned}
$$

Thus, the total angular momentum about the symmetry axis $L_z = $ constant

## 3 Bead on a Spinning Hoop

$D$ is the distance between point $A$ and the particle.
$R$ is the radius of the hoop.
We have the following relations for the system:

$$
\begin{aligned}
    \dot{\theta} &= \omega \\
    \alpha &= \theta + \frac{\phi}{2}\\

\end{aligned}
$$

### Question a

In Cartesian coordinates:

$$
\begin{aligned}
    x &= D\cos \alpha\\
    y &= D\sin \alpha\\
    \dot{x} &= \dot{D}\cos\alpha - D\dot{\alpha}\sin\alpha\\
    \dot{y} &= \dot{D}\sin\alpha + D\dot{\alpha}\cos\alpha\\
\end{aligned}   
$$

Kinetic energy:

$$
\begin{aligned}
    T &= \frac{1}{2}m(\dot{x}^2 + \dot{y}^2)\\\\
    &= \frac{1}{2}m((\dot{D}\cos\alpha - D\dot{\alpha}\sin\alpha)^2 + (\dot{D}\sin\alpha + D\dot{\alpha}\cos\alpha)^2)\\\\ 
    &= \frac{1}{2}m(\dot{D}^2\cos^2\alpha + \dot{D}^2\sin^2\alpha +D^2\dot{a}^2\sin^2\alpha + D^2\dot{a}^2\cos^2\alpha)\\\\
    &= \frac{1}{2}m(\dot{D}^2 + D^2\dot{\alpha}^2)
\end{aligned}
$$

Given

$$
\begin{aligned}
    \dot{D} &= -R\dot{\phi}\sin(\frac{\phi}{2})\\\\
    \dot{\alpha} &= \omega + \frac{\dot{\phi}}{2}
\end{aligned}
$$

We have

$$
\begin{aligned}
    T &= \frac{1}{2}m(R^2\dot{\phi}^2\sin^2(\frac{\phi}{2}) + 4R^2\cos^2(\frac{\phi}{2})(\omega + \frac{\dot{\phi}}{2})^2)
\end{aligned}
$$

Since the motion is horizontal, there is no change in potential. Thus, 

$$
\mathcal{L} = T - U = T = \frac{1}{2}m(R^2\dot{\phi}^2\sin^2(\frac{\phi}{2}) + 4R^2\cos^2(\frac{\phi}{2})(\omega + \frac{\dot{\phi}}{2})^2)
$$

### Question b

$$
\begin{aligned}
    \frac{\partial\mathcal{L}}{\partial\phi} &= \frac{1}{2}mR^2(\dot{\phi}^2\sin(\frac{\phi}{2})\cos(\frac{\phi}{2}) - 4\sin(\frac{\phi}{2})\cos(\frac{\phi}{2})(\omega + \frac{\dot{\phi}}{2})^2) \\\\
    &= \frac{1}{2}mR^2(\frac{1}{2}\dot{\phi}^2\sin\phi - 2\sin\phi(\omega^2 + \omega\dot{\phi}+\frac{\dot{\phi}^2}{2}))\\\\
    &= -mR^2\sin\phi(\omega^2+\omega\dot{\phi})\\\\\\
    \frac{\partial\mathcal{L}}{\partial\dot{\phi}} 
    &= \frac{1}{2}mR^2(2\dot{\phi}\sin^2(\frac{\phi}{2}) +4\cos^2(\frac{\phi}{2})(\omega + \frac{\phi}{2})) \\\\
    &=  mR^2(\dot{\phi} + 2\omega\cos^2(\frac{\phi}{2})) \\\\
    \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\phi}} 
    &= mR^2(\ddot{\phi} - \omega\dot{\phi}\sin\phi) \\\\
\end{aligned}
$$

According to Euler-Lagrange equation for $\phi$, $\frac{\partial\mathcal{L}}{\partial\phi} = \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\phi}}$, thus we have:

$$
\begin{aligned}
    -mR^2\sin\phi(\omega^2+\omega\dot{\phi}) &= mR^2(\ddot{\phi} - \omega\dot{\phi}\sin\phi) \\\\
    -\omega^2\sin\phi - \omega\dot{\phi}\sin\phi &= \ddot{\phi} - \omega\dot{\phi}\sin\phi \\\\
    \ddot{\phi} = -\omega^2 \sin\phi
\end{aligned}
$$

which is exactly the equation of motion of a simple pendulum

With a very small amplitude, $\sin\phi\approx \phi$, thus:

$$
\ddot{\phi} = -\omega^2\phi
$$

which is the equation of motion of a simple harmonic oscillator with angular frequency $\omega$.

## 4 Particle in a Spherical Bowl

### Question a
$$
\begin{aligned}
    T &= \frac{1}{2}mR^2(\dot{\theta}^2 + \sin^2\theta\dot{\phi}^2) \\\\
    U &= -mgR\cos\theta \\\\
    \mathcal{L} &= T - U = \frac{1}{2}mR^2(\dot{\theta}^2 + \sin\theta^2\dot{\phi}^2) + mgR\cos\theta
\end{aligned}
$$

### Question b

Euler-Lagrange equation for $\phi$ coordinate:

$$
\begin{aligned}
    \frac{\partial \mathcal{L}}{\partial\phi} - \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\phi}} &= 0 -\frac{d}{dt}(mR^2\sin^2\theta\dot{\phi}) = 0 \\\\
    0 &= \frac{d}{dt}(mR^2\sin^2\theta\dot{\phi}) \\\\
    &= \frac{d}{dt}((R\sin\theta)(mR\sin\theta\dot{\phi})) \\\\
    &= \frac{d}{dt}|\vec{r}\times\vec{p}| \\\\
    &= \frac{d}{dt}l_z
\end{aligned}
$$

Thus, $l_z = $ constant.

### Question c

Euler-Lagrange euqation for $\theta$ coordinate:

$$
\begin{aligned}
    \frac{\partial \mathcal{L}}{\partial\theta} 
    & = mR^2\sin\theta\cos\theta\dot{\phi}^2 -mgR\sin\theta \\\\
    \frac{\partial\mathcal{L}}{\partial\dot{\theta}} 
    & = mR^2\dot{\theta} \\\\
    \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\theta}} 
    &= mR^2\ddot{\theta} \\\\
    0 &= \frac{\partial \mathcal{L}}{\partial\theta} - \frac{d}{dt}\frac{\partial\mathcal{L}}{\partial\dot{\theta}} 
    = mR^2\sin\theta\cos\theta\dot{\phi}^2 -mgR\sin\theta - mR^2\ddot{\theta} \\\\
\end{aligned}
$$

Substituting $\dot{\phi} = \frac{l_z}{mR^2\sin^2\theta}$ into the equation we have the equation of motion involving only $\theta$:

$$
0 = \ddot{\theta} - \frac{l_z^2}{m^2R^4}\frac{\cos\theta}{\sin^3\theta} + \frac{g}{R}\sin\theta
$$

### Question d

Substituting $\ddot{\theta} = \frac{d\dot{\theta}}{d\theta}\dot{\theta}$ into the equation of motion we get:

$$
\begin{aligned}
    0 &= \frac{d\dot{\theta}}{d\theta}\dot{\theta} - \frac{l_z^2}{m^2R^4}\frac{\cos\theta}{\sin^3\theta} + \frac{g}{R}\sin\theta \\\\
    \dot{\theta}d\dot{\theta} &=(\frac{l_z^2}{m^2R^4}\frac{\cos\theta}{\sin^3\theta} - \frac{g}{R}\sin\theta)d\theta
\end{aligned}
$$

Integrating both sides:

$$
\frac{1}{2}\dot{\theta}^2 = \int(\frac{l_z^2}{m^2R^4}\frac{\cos\theta}{\sin^3\theta} - \frac{g}{R}\sin\theta)d\theta
$$

Let $u = \sin\theta$, then $\frac{du}{d\theta} = \cos\theta$, then 

$$
\begin{aligned}
    \frac{1}{2}\dot{\theta}^2 &= \frac{l_z^2}{m^2R^4}\int\frac{1}{u^3}du - \frac{g}{R}\int\sin\theta d\theta \\\\
    \frac{1}{2}\dot{\theta}^2 &= -\frac{1}{2}\frac{l_z^2}{m^2R^4}\frac{1}{\sin^2\theta} + \frac{g}{R}\cos\theta + \epsilon \\\\
    \epsilon &= \frac{1}{2}\dot{\theta}^2 + \frac{1}{2}\frac{l_z^2}{m^2R^4}\frac{1}{\sin^2\theta} - \frac{g}{R}\cos\theta
\end{aligned}
$$

### Question e

```python
import numpy as np 
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from collections import deque
```

```python
%matplotlib notebook
```

```python
# radius of bowl, g, mass of particle, 
# initial phi angular velocity, initial theta angle, angular momentum
R = 10 # m 
g = 9.81 # m/(s^2)
m = 1 # kg
phi_dot_0 = 0.2*np.pi # rad/s
theta_0 = 0.4*np.pi # rad
l_z = m * (R**2) * (np.sin(theta_0)**2) * phi_dot_0 # kg * m^2 / s

# convinience cosntants
omega_sqr = g/R
a_sqr = (l_z**2) / (m**2 * R**4)
epsilon = a_sqr / (2*np.sin(theta_0)**2) - omega_sqr*np.cos(theta_0)


# initial condition for trajectory plot
w0  = 0, phi_dot_0, theta_0, 0
t_min = 0
t_max = 30
t = np.linspace(t_min, t_max, 1000)
print(l_z)
print(epsilon)
```

```python
# Effective Potential Plot
theta = np.linspace(0., np.pi/2, 1000)
u_eff = a_sqr / (2*np.sin(theta)**2) - omega_sqr*np.cos(theta)

plt.figure()
plt.plot(theta, u_eff, label=r'$U_{eff}$')
plt.axhline(epsilon, linestyle='--', label=r'$\epsilon$')
plt.xlabel(r'$\theta$')
plt.ylabel('Energy')
plt.ylim(np.min(u_eff)-0.1, u_eff[-1])
plt.legend()
```

```python
# Helper functions for trajectory plot

# system of ODE's as EoM of the particle
def trajectory(t, w):
    
    # positions, angular velocities
    phi, phi_dot, theta, theta_dot = w
    
    # angular accelerations
    phi_dotdot = - phi_dot * theta_dot * np.arctan(theta)
    theta_dotdot = a_sqr * np.cos(theta) / np.sin(theta)**3 - omega_sqr * np.sin(theta)
    
    return phi_dot, phi_dotdot, theta_dot, theta_dotdot

# converts Sphereical Coordinates to Cartesian Coordinates
def sph2cart(phi, theta, r):
    rho = r * np.sin(theta)
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    z = r * np.cos(theta)
    return x, y , z
```

```python
# Trajectory Plot

sol = solve_ivp(lambda t, w: trajectory(t, w),
                (t_min, t_max),
                w0,
                dense_output = True)

# the particle's trajectory
coords = sol.sol(t)
u, v= coords[0], coords[2]
x = sph2cart(u, v, R)[0]
y = sph2cart(u, v, R)[1]
z = -sph2cart(u, v, R)[2]

# the bowl's surface
u, v = np.mgrid[0:2 * np.pi:30j, 0.5*np.pi:np.pi:30j]
x_bowl = sph2cart(u, v, R)[0]
y_bowl = sph2cart(u, v, R)[1]
z_bowl = sph2cart(u, v, R)[2]

# 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.locator_params(nbins = 5)
ax.set_xlabel("(m)")
ax.set_ylabel("(m)")
ax.set_zlabel("(m)")
# plotting the bowl
ax.plot_surface(x_bowl, y_bowl, z_bowl, color='blue', alpha = 0.1)
# plotting the trajectory
ax.plot(x, y, z, color='red')
plt.show()
```

#### Qualitative Description

While going around about the $z$ axis, the particle also wobbles periodically up and down the side of the bowl, never reaching the bottom. 

Given the initial conditions that $\theta_0 = 0.4\pi$ and $\dot\phi_0 = 0.2\pi\space s^{-1}$, 

At the turning point where $\theta \approx 0.49$, $\dot{\theta} = 0$, the particle is at its lowest verticle position in the bowl.

At the turning point where $\theta \approx 1.26$ and $\dot{\theta} = 0$, the particle is at its highest verticle position in the bowl.

There is also a stationary value of $\theta \approx 0.77$, where, given the right initial condition, $\theta$ will remain constant and the particle only has the motion of rotating about the $z$ axis.
