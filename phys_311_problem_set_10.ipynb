## Particle in a Spherical Bowl Revisited

Solving for $p_\theta$ we have:

$$
p_{\theta} = \sqrt{mR^2 E + m^2 R^3g\cos\theta - \frac{l_z^2}{\sin^2\theta}}
$$

However, instead of plotting a value-to-value graph of $p_{\theta}$ and $\theta$, it is easier to plot a contour of the total energy on the $\theta$ - $p_{\theta}$ plane using the equation:

$$
E = \frac{1}{2}\frac{p_{\theta}^2}{mR^2} + \frac{1}{2}\frac{l_z^2}{mR^2\sin^2\theta} - mR^2\frac{g}{R}\cos\theta
$$


```python
R = 1 # m 
g = 9.81 # m/(s^2)
m = 1 # kg
l_z = 1 # taken from Problem Set 6


theta = np.linspace(0., np.pi, 1000)
p_theta = np.linspace(-100, 100, 1000)

tt, pp = np.meshgrid(theta, p_theta)

EE = (0.5/m/R**2)*pp**2 + 0.5*l_z**2/m/R**2/np.sin(tt)**2 - m*R*g*np.cos(tt)

levels = [-6, -2, 6, 10, 20]

#m_p_theta = -np.sqrt(m*R**2*E + m**2*R**3*g*np.cos(theta) - l_z**2/np.sin(theta)**2)
#p_p_theta = np.sqrt(m*R**2*E + m**2*R**3*g*np.cos(theta) - l_z**2/np.sin(theta)**2)

plt.figure()
plt.title('$R=1\:m$\n$g=9.81\:m\cdot s^{-2}$\n$m=1\:kg$\n$\ell_z=1\:(kg\cdot m^2\cdot s^{-1})$', loc='left')
contourplot = plt.contour(tt, pp, EE, levels=levels)
plt.xlabel(r'$\theta\: (rad)$')
plt.ylabel(r'$p_{\theta}\: (kg\cdot m^2\cdot s^{-1})$')
plt.ylim(-10,10)
plt.tight_layout()
plt.legend(handles=contourplot.collections, labels=[r'$E=-6\: J$',
                                                    r'$E=-2\: J$', 
                                                    r'$E=6\: J$', 
                                                    r'$E=10\: J$', 
                                                    r'$E=20\: J$'])
```
